```python
# Merge csv files.

import glob

def hebing():
    csv_list = glob.glob('*.csv')
    print(u'Find %s csv files'% len(csv_list))
    print(u'Processing............')
    
    for i in csv_list:
        fr = open(i,'r').read()
        with open('allimdb.csv','a') as f:
            f.write(fr)
    print(u'Done！')

if __name__ == '__main__':
    hebing()
```

```python
import re 
import collections 
import numpy as np 
import jieba 

fn = open('title3.txt')
string_data = fn.read()
fn.close() 

pattern = re.compile(u'\t|\n|\.|-|:|;|\)|\(|\?|"') 
string_data = re.sub(pattern, '', string_data) 

seg_list_exact = jieba.cut(string_data, cut_all = False) 
object_list = []
remove_words = [u'the', u'a',u'of', u'in', u'to', u'de', u'la',u'and',u'el',u'la',u'vs',u'el',u'on',u'and',u'for',u'da',
                u'by',u'at',u'is',u' ']

for word in seg_list_exact:
    if word not in remove_words:
        object_list.append(word)


word_counts = collections.Counter(object_list)
word_counts_top100 = word_counts.most_common(100) 
word_counts_top100
```

```python
import re 
import collections 
import numpy as np 
import jieba 

fn = open('genre2.txt')
string_data = fn.read()
fn.close() 

pattern = re.compile(u'\t|\n|\.|-|:|;|\)|\(|\?|"') 
string_data = re.sub(pattern, '', string_data) 

seg_list_exact = jieba.cut(string_data, cut_all = False) 
object_list = []
remove_words = [u'genre',u' ']

for word in seg_list_exact:
    if word not in remove_words:
        object_list.append(word)


word_counts2 = collections.Counter(object_list)
word_counts2_top20 = word_counts2.most_common(20) 
word_counts2_top20
```
