```python
#coding=utf-8
import requests
import json
from pyquery import PyQuery as pq
import csv
import os
headers = {'user-agent': 'Mozilla/5.0 (iPhone; CPU iPhone OS 11_0 like Mac OS X) AppleWebKit/604.1.38 (KHTML, like Gecko) Version/11.0 Mobile/15A372 Safari/604.1'}

def fetch_movie_detail(url):
    r = requests.get(url)
    doc = pq(r.text)
    info = doc('div #info').text().split('\n')
    try:
        # print(info[3])
        # print(info[4])
        # print(info[6])
        return info[3].split(':')[1], info[4].split(':')[1], info[6].split(':')[1]
    except:
        print(info)
        return '', '', ''

def load_movie_list():
    start = 0 
    end = 3000
    internal = 10
    base_url = 'https://movie.douban.com/j/new_search_subjects?sort=U&range=2,4&tags=电影&start={0}&year_range=2010,2019'
    movie_list = []
    for index in range(start, end, internal):
        url = base_url.format(index)
        r = requests.get(url)
        if r.status_code == 200:
            movies = json.loads(r.text)
            # print(movies['data'])
            for movie in movies['data']:
                title = movie['title']
                directors = movie['directors']
                rate = movie['rate']
                movie_type, producer, publish_date = fetch_movie_detail(movie['url'])
                print('')
                print('title', movie['title'])
                print('directors', *movie['directors'])
                print('rate', movie['rate'])
                print('type', movie_type)
                print('producer', producer)
                print('publish_date', publish_date)
                movie_list.append([title, directors, rate, movie_type, producer, publish_date])
    write_to_csv(movie_list)
            
def write_to_csv(movie_list):
    headers = ['title', 'directors', 'rate', 'type', 'producer', 'publish_date']
    path = '/Users/suosuo/Downloads/douban.csv'
    if os.path.exists(path):
        os.remove(path)
    with open(path, 'w') as f:
        f_csv = csv.writer(f)
        f_csv.writerow(headers)
        for item in movie_list:
            f_csv.writerow(item)
        f.flush()

load_movie_list()
```
