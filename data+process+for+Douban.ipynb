```python
data_new = data.copy()
```

```python
content_type = [' 喜剧', ' 动作 ', ' 剧情 ', ' 喜剧 ', ' 爱情 ', ' 爱情', ' 动画 ',
       ' 恐怖', ' 悬疑 ', ' 科幻', ' 惊悚 ', ' 剧情', ' 科幻 ', ' 奇幻 ', ' 惊悚', ' 情色',
       ' 美国', ' 动作', ' 韩国', ' 歌舞', ' 泰国', ' 儿童 ',
       ' 韩语', ' 恐怖 ', ' 武侠 ', ' 汉语普通话', ' 历史 ', ' 英语', ' 动画', ' 日本',
       ' 台湾', ' 美国 ', ' 战争 ', ' 悬疑', ' 古装']
```

```python
data_new = data_new[data_new.type.isin(content_type)]
```

```python
data_new = data_new.reset_index(drop=True)
```

最后的出版日期列只保留年份

```python
import re
pattern = re.compile(r'\d+')
for i in range(len(data_new)):
    if data_new['publish_date'][i][1] == str(2):
        data_new['publish_date'][i] = pattern.findall(data_new['publish_date'][i])[0]
# result1 = pattern.findall(data['publish_date'][1])[0]
```

```python
final_data = data_new[data_new.publish_date.isin(['2013','2014','2015','2016','2017'])]
```

```python
final_data = final_data.reset_index(drop=True)
```
